```python
import pandas as pd

f_sen = pd.read_csv("factors_sen.csv")
f_jedzenie = pd.read_csv("factors_jedzenie.csv")
l_sen = pd.read_csv("loadings_sen.csv")
l_jedzenie = pd.read_csv("loadings_jedzenie.csv")
f_sen.columns = ["index", "Factor_1", "Factor_2", "Factor_3"]
f_jedzenie.columns = ["index", "Factor_1", "Factor_2", "Factor_3"]
l_sen.columns = ["index", "Factor_1", "Factor_2", "Factor_3"]
l_jedzenie.columns = ["index", "Factor_1", "Factor_2", "Factor_3"]
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Assuming f_sen and f_jedzenie are DataFrames with the relevant columns
factors_sen = f_sen[["Factor_1", "Factor_2", "Factor_3"]].values
factors_jedzenie = f_jedzenie[["Factor_1", "Factor_2", "Factor_3"]].values

# Compute the Spearman correlation matrix using scipy
corr_matrix = np.zeros((3, 3))
p_vals = np.zeros((3, 3))
for i in range(3):
    for j in range(3):
        corr_matrix[i, j], p_vals[i, j] = spearmanr(factors_sen[:, i], factors_jedzenie[:, j])

# Plot the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", xticklabels=["Factor_1", "Factor_2", "Factor_3"], yticklabels=["Factor_1", "Factor_2", "Factor_3"])
plt.show()
print(p_vals)
```
