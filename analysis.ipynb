```python
import pandas as pd
import numpy as np
from scipy.stats import pointbiserialr

# Clean numeric data
df_clean_num = num_data.dropna()
# Compute Spearman correlation for numerical data
corr_num = df_clean_num.corr(method='spearman')

# Clean categorical (dummy) data
df_clean_dum = dum_data.dropna()
# Compute Pearson correlation for categorical (dummy) data
corr_dum = df_clean_dum.corr(method='pearson')

# Convert Boolean to Float (True=1, False=0) while keeping NaNs intact
df_bool = dum_data.astype(float)

# Create empty DataFrames for correlations and p-values
corr_matrix = pd.DataFrame(index=df_bool.columns, columns=num_data.columns, dtype=float)
p_values_matrix = pd.DataFrame(index=df_bool.columns, columns=num_data.columns, dtype=float)

# Compute Point-Biserial Correlation (handling NaNs)
for bool_col in df_bool.columns:
    for num_col in num_data.columns:
        # Drop NaNs in both columns before computing correlation
        df_pair = df_bool[[bool_col]].join(num_data[[num_col]]).dropna()

        if len(df_pair) > 1:  # Ensure at least 2 valid observations
            corr, p_value = pointbiserialr(df_pair[bool_col], df_pair[num_col])
            p_values_matrix.loc[bool_col, num_col] = p_value
            corr_matrix.loc[bool_col, num_col] = corr

# Copy the computed correlation matrix
corr_dum_num = corr_matrix.copy()
```

```python
import numpy as np
from scipy.stats import norm

def fisher_z(r):
    return 0.5 * np.log((1 + r) / (1 - r))

def compare_correlations(r1, r2, n1, n2):
    z1 = fisher_z(r1)
    z2 = fisher_z(r2)
    se = np.sqrt(1 / (n1 - 3) + 1 / (n2 - 3))
    z = (z1 - z2) / se
    p = 2 * (1 - norm.cdf(abs(z)))  # two-tailed
    return z, p

# Example:
r1 = corr_matrix[0, 1]   # correlation 1
r2 = corr_matrix[1, 1]   # correlation 2
n1 = len(f_sen)
n2 = len(f_sen)

z, p = compare_correlations(r1, r2, n1, n2)

print(f"z = {z:.3f}, p = {p:.3f}")
```
